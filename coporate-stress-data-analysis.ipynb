```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
corporate_df = pd.read_csv("corporate_stress_dataset.csv")
corporate_df
```

```python
plt.figure(figsize = (8,6))
sns.histplot(corporate_df["Stress_Level"], kde = True)
plt.title("Distribution of Stress Level")
plt.show()
```

```python
numerical_features = corporate_df.select_dtypes(include=np.number).columns.tolist()

print(numerical_features)
```

```python
num_cols = 3
num_rows = int(len(numerical_features) / num_cols) + (1 if len(numerical_features) % num_cols != 0 else 0)
plt.figure(figsize=(20, 6*num_rows))

for index, feature in enumerate(numerical_features):
    plt.subplot(num_rows, num_cols, index + 1)
    sns.histplot(corporate_df[feature], kde=True, bins=30)
    plt.xlabel(feature, fontsize=14)
    plt.ylabel('Count', fontsize=14)
    plt.title(f"Distribution of {feature}", fontsize=16)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.tight_layout(rect=[0, 0, 1, 0.95])

plt.show()
```

```python
plt.figure(figsize=(20, 6*num_rows))
for index, feature in enumerate(numerical_features):
    plt.subplot(num_rows, num_cols, index + 1)
    sns.boxplot(y=corporate_df[feature])
    plt.title(f"Box plot of {feature}", fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.95])

plt.show()
```

```python
categorical_features = corporate_df.select_dtypes(exclude=np.number).columns.tolist()
print(categorical_features)
plt.figure(figsize=(20, 6*num_rows))
for index, feature in enumerate(categorical_features):
    plt.subplot(num_rows, num_cols, index + 1)
    sns.countplot(x=feature, data=corporate_df, order=corporate_df[feature].value_counts().index)
    plt.title(f'Distribution of {feature}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout(rect=[0, 0, 1, 0.95])

plt.show()
```

```python
plt.figure(figsize=(20, 6*num_rows))

for index, feature in enumerate(numerical_features):
    plt.subplot(num_rows, num_cols, index + 1)
    sns.scatterplot(x=feature, y='Stress_Level', data=corporate_df)
    plt.title(f'Scatter Plot of {feature} vs. Stress Level')

plt.show()
```

```python
plt.figure(figsize=(20, 6*num_rows))

for index, feature in enumerate(categorical_features):
    plt.subplot(num_rows, num_cols, index + 1)
    sns.boxplot(x=feature, y='Stress_Level', data=corporate_df)
    plt.title(f'Scatter Plot of {feature} vs. Stress Level')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()

plt.show()
```

```python
correlation_matrix = corporate_df[numerical_features].corr()

plt.figure(figsize=(20, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
